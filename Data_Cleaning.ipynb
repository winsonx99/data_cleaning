## Check the content of the column- ‘Title’ and define a function to clean the value. E.g., the original value: Walter Forbes. [A novel.] By A. A to: Walter Forbes;  Love the Avenger. By the author of “All for Gr.. to Love The Avenger.

```python
def clean_title(title):

    if title == 'nan':
        return 'NaN'

    if '.' in title:
        title = title[:title.find('.')]
        # title.find('.') <- get the index of the first '.'
        # title[:title.find('.')] <- trim title and keep only those char on the left of '.'
    if title[0] == '[':
        title = title[1: title.find(']')]
    if '[' in title:
        title = title[:title.find('[')]

    if 'by' in title:
        title = title[:title.find('by')]
    elif 'By' in title:
        title = title[:title.find('By')]


    title = title[:-2]

    # Capatlize each word by cut them apart, capitalize each item and merge them back together
    title = list(map(str.capitalize, title.split()))
    return ' '.join(title)

df2['Title cleaned'] = df2['Title'].apply(clean_title)
df2.head()
```
